```python
#especialidade, médico, clinica, tempo_espera

df_consultas_novo = df_consultas[['especialidade', 'id_medico', 'id_clinica', 'tempo_espera']]
df_consultas_novo
```

```python
#quero o percentual de consultas em relaçao ao total de consultas realizadas cujo atraso do atendimento demorou mais do que 20 minutos, POWER BI

#encontrar o total de consultas cujo tempo foi maior que 20 minutos
df_tempo_espera_maior = df_consultas_novo['tempo_espera'] > pd.Timedelta(minutes = 20)
#df_total_tempo_espera_maior = df_consultas_novo.loc[df_tempo_espera_maior].shape[0]

df_tempo_espera_maior
```

```python
#PEDIDOS DO PROJETO, identificar outliers usando quartis 
#tempo_espera
#pegar tempo de espera, organizar de forma crescente, separar em 4 pedaços e ver quais são as métricas baseado no calculo dos outliers 
#IQR = Q3 - Q1
#Limite superior = Q3 + (1.5 × IQR)
#Limite inferior = Q1 - (1.5 × IQR)

array_tempo_espera = np.array(df_consultas_novo['tempo_espera'])
array_tempo_espera

#converter o formato para timedelta minutos
array_tempo_espera_minutos = array_tempo_espera.astype('timedelta64[m]')
array_tempo_espera_minutos

```
